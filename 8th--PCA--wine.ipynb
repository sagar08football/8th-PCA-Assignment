```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine = pd.read_csv("C:/Users/Narsing/Downloads/Assigments data science/8th--Assigment--PCA/wine.csv")
wine
```

```python
WINE = wine.values
WINE
```

```python
wine_normal = scale(WINE)
```

```python
wine_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
```

```python
pc = pd.DataFrame(pca_values)
pc
```

```python
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2'])], axis = 1)
```

```python
finalDf
```

```python

from sklearn.cluster import KMeans
```

```python
wcss1 = []

kmeans = KMeans(n_clusters=3,random_state=0)
kmeans.fit(finalDf)
wcss1.append(kmeans.inertia_)
```

```python
wcss1
```

```python
kmeans1 = KMeans(n_clusters=4,random_state=0)

kmeans1.fit(finalDf)

wcss1.append(kmeans1.inertia_)
```

```python
wcss1
```

```python
Pcadf = pd.DataFrame(finalDf)
```

```python
Pcadf['Cluster_3'] = kmeans.labels_
```

```python
Pcadf.head()
```

```python
wcss = []
for i in range(1, 16):
    kmeans = KMeans(n_clusters=i,random_state=145)
    kmeans.fit(finalDf)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 16), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
